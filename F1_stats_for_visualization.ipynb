Okay, so we see that we have all the information we wanted except from the date. Let's see if we can find that in the races dataset.

```python
#races dataset
races = pd.read_csv('./races.csv', encoding = "latin")
races.head()
```

We have it (not as a date, but that is okay for now)!!!

Let's now select the columns we are interested in before we join all of them into one dataset.

```python
#selecting the columns of interest in the circuits dataset
circuits = circuits[['circuitId', 'name', 'country', 'lat', 'lng']]
```

```python
#selecting the columns of interest in the drivers dataset
drivers = drivers[['driverId','forename', 'surname',
       'nationality']]
```

```python
#selecting the columns of interest in the races dataset
races = races[['raceId', 'circuitId', 'date']]
```

```python
#selecting the columns of interest in the winners dataset
winners = winners[['raceId', 'driverId', 'grid', 'points']]
```

We have now made the dataset smaller with the information we want. Let's join them!!

```python
winner_races = pd.merge(winners, races, on='raceId', how='inner')
```

```python
winner_races_drivers = pd.merge(winner_races, drivers, on='driverId', how='inner')
```

```python
race_winners = pd.merge(winner_races_drivers, circuits, on='circuitId', how='inner')
```

```python
race_winners.head(20)
```

```python
len(race_winners) #it makes sense, because it is the same lenght as the lenght of the winners dataset
```

```python
#let's check for null values
race_winners.info()
```

```python
race_winners.columns
```

```python
#let's get rid of the Id columns that we don't need and put the dataset in a nice order
race_winners = race_winners[['date', 'forename', 'surname', 'nationality', 'name', 'country', 'grid', 'points', 'lat', 'lng']]
```
